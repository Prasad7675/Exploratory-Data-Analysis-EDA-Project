```python
import pandas as pd

data = {
    "Name": ["Rahul","Priya","Amit","Sneha","Rohan",
             "Pooja","Karan","Neha","Arjun","Anjali"],
    "Gender": ["Male","Female","Male","Female","Male",
               "Female","Male","Female","Male","Female"],
    "Maths": [78,88,67,90,55,76,80,85,72,91],
    "Science": [85,92,70,95,60,82,78,89,74,94],
    "English": [80,95,65,93,58,79,84,87,70,96],
    "Attendance": [90,96,85,98,75,88,91,94,82,99]
}

df = pd.DataFrame(data)

df.to_csv("eda_students.csv", index=False)

print("Dataset created successfully")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("eda_students.csv")

# Display dataset
print("Dataset:")
print(df)

# Dataset information
print("\nDataset Info:")
print(df.info())

# Statistical summary
print("\nStatistical Summary:")
print(df.describe())

# Create Average Marks column
df["Average"] = (df["Maths"] + df["Science"] + df["English"]) / 3

# Highest scoring student
top_student = df.loc[df["Average"].idxmax()]
print("\nTop Performing Student:")
print(top_student)

# Bar Chart
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Average"])
plt.title("Average Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Average Marks")
plt.xticks(rotation=45)
plt.show()

# Histogram
plt.figure(figsize=(7,5))
plt.hist(df["Attendance"], bins=5)
plt.title("Attendance Distribution")
plt.xlabel("Attendance")
plt.ylabel("Number of Students")
plt.show()

# Scatter Plot
plt.figure(figsize=(7,5))
plt.scatter(df["Attendance"], df["Average"])
plt.title("Attendance vs Average Marks")
plt.xlabel("Attendance")
plt.ylabel("Average Marks")
plt.show()

# Heatmap
sns.heatmap(df[["Maths","Science","English","Attendance","Average"]].corr(),
            annot=True)

plt.title("Correlation Heatmap")
plt.show()
```
